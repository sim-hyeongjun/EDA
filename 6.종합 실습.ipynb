```python
tips= sns.load_dataset('tips')
print(tips)
fig= plt.figure()
axes1=fig.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('bill vs tip')
axes1.set_ylabel('total_bill')
axes1.set_xlabel('tip')
plt.show()
```

```python
fig =plt.figure()
axes1= fig.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
             tips[tips['sex']=='Male']['tip']],labels=['female', 'male'])
axes1.set_title('boxplot of tip by sex')
axes1.set_xlabel('sex')
axes1.set_ylabel('tips')
```

```python
def recod_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1
tips['sex_color']= tips['sex'].apply(recod_sex)
fig=plt.figure()
axes1= fig.add_subplot(1,1,1)
axes1.scatter(x=tips['total_bill'], y=tips['tip'], s=tips['size']*10, c=tips['sex_color'],alpha=0.5)
axes1.set_title('Scatter of Total Bill vs Tip Color by Sex and Sized by Size' )
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
ax= plt.subplot()
ax=sns.displot(tips['total_bill'],kde=False)
axes1.set_title('Scatter of Total desity' )
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('frequency')
```

```python
ax= plt.subplot()
ax=sns.displot(tips['total_bill'],kde=True)
axes1.set_title('Scatter of Total desity' )
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('frequency')
```

```python
ax= plt.subplots()
ax=sns.displot(tips['total_bill'])
axes1.set_title('Scatter of Total desity' )
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('unit pro')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True) #rug인수는 그래프의 축에 동일한 길이의 직선을 붙여서 데이터의 밀집 정도를 표현함
ax.set_title('Total Bill Histogram with Density and Rug Plot ' )
ax.set_xlabel('Total Bill')

```

```python
#countplot() : 이산값 표현 그래프
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
#regplot() : 산점도 + 회귀선 , fit_reg인자를 False로 설정하면 회귀선을 제거
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatter & Reg of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatter & Reg of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#jointplot() : 산점도 + 히스토그램을 함께 표현
fig = plt.figure()
ax = sns.jointplot(x='total_bill', y='tip', data=tips)
ax.fig.suptitle('Jointplot of Total Bill and Tip', fontsize=10, y=1.03)
ax.set_axis_labels(xlabel='Total Bill', ylabel ='Tip' )
```

```python
fig = plt.figure()
ax = sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex") #육각형으로 데이터를 쌓아 표현하는 그래프
ax.fig.suptitle('Jointplot of Total Bill and Tip', fontsize=10, y=1.03)
ax.set_axis_labels(xlabel='Total Bill', ylabel ='Tip' )
```

```python


#kdeplot() : 이차원 밀집도 표현 그래프
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True) #shade는 음영효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#barplot() : 지정한 변수의 평균을 계산해서 그래프에 표현 가능
#시간에 따라 지불한 비용의 평균을 barplot 으로 표현
ax = plt.subplots()
ax = sns.barplot(x = 'time', y='total_bill' , data =tips)
ax.set_title('Bar Plot of average Total Bill for time of day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Average Total Bill')
```
